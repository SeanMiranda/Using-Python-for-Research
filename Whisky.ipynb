```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
whisky = pd.read_csv("whiskies.txt")
whisky['Region'] = pd.read_csv('regions.txt')
```

```python
flavors = whisky.iloc[:,2:14]
flavors
```

```python
corr_flavors = pd.DataFrame.corr(flavors)
corr_flavors
```

```python
plt.figure(figsize=(10,10))
plt.pcolor(corr_flavors)
plt.colorbar()
plt.savefig('corr_flavor.pdf')
```

```python
# Each whisky is from a different distillery so we get the flavor attributes for each distillery
flavors.transpose()
```

```python
corr_whisky = pd.DataFrame.corr(flavors.transpose())
plt.figure(figsize=(10,10))
plt.pcolor(corr_whisky)
plt.colorbar()
plt.axis('tight')
plt.savefig('corr_whisky.pdf')
```

```python
from sklearn.cluster.bicluster import SpectralCoclustering
```

```python
model = SpectralCoclustering(n_clusters=6, random_state=0)
model.fit(corr_whisky)
```

```python
model.rows_
```

```python
# How many whiskies are in each of the 6 clusters (6 regions)
np.sum(model.rows_, axis=1)
```

```python
# The sum should be 1 because each whisky belongs to 1 cluster 
np.sum(model.rows_, axis=0)
```

```python
# observation 0 belongs to cluster 2 ... 
model.row_labels_
```

```python
whisky['Group'] = pd.Series(model.row_labels_, index=whisky.index)
whisky = whisky.iloc[np.argsort(model.row_labels_)]
whisky = whisky.reset_index(drop=True)
```

```python
whisky
```

```python
correlations = pd.DataFrame.corr(whisky.iloc[:,2:14].transpose())
correlations = np.array(correlations)
```

```python
plt.figure(figsize= (14,7))
plt.subplot(121)
plt.pcolor(corr_whisky)
plt.title('Original')
plt.axis('tight')
plt.subplot(122)
plt.pcolor(correlations)
plt.title('Rearranged after clustering into 6 regions')
plt.axis('tight')
plt.savefig('correlations.pdf')
```
